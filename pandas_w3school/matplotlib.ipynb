## Plot values on (x,y) graph

```python
import matplotlib.pyplot as plt
import numpy as np

xpoint = np.array([0,6])
ypoint = np.array([0,250])

plt.plot(xpoint,ypoint)
plt.show()
```

```python
xpoint = np.array([1,3,5,7])
ypoint = np.array([50,175,120,150])

# plt.plot(xpoint,ypoint)
plt.plot(xpoint,ypoint,'o')
plt.show()
```
