### Data structure: [restaurant_id, 2021, 2022, 2023, 2024]

```python
import numpy as np
import matplotlib.pyplot as plt

sales_data = np.array([
[1, 150000, 180000, 220000, 250000], # Paradise Biryani
[2, 120000, 140000, 160000, 190000], # Beijing Bites
[3, 200000, 230000, 260000, 300000], # Pizza Hub
[4, 180000, 210000, 240000, 270000], # Burger Point
[5, 160000, 185000, 205000, 230000] # Chai Point
])


print("==== Sales Alalysis ====")
print(sales_data.shape)
print("1st 3 restraunt:",sales_data[:3],sep="\n")
```

## Total sales per year

```python
print("Total sales per year wise for all items / column:",np.sum(sales_data,axis=0))
print("Total sales per year wise for single item/ row:",np.sum(sales_data[:,1:],axis=1))

# Minimum sales per restaurant
print("Minimum sales per item", np.min(sales_data[:,1:],axis=1))

#Maximum sales per year 
print("Max sales per item", np.max(sales_data[:,1:],axis=0))

#Average sales per restaurant
print("Avg sales per item", (np.mean(sales_data[:,1:],axis=1)))

#cumlative sales per restaurant
cumulative = (np.cumsum(sales_data[:,1:],axis=1))
print("CumSum sales per item",cumulative)
```

## Plot Matplotlib

```python
plt.figure(figsize=(8,6))
plt.plot(np.mean(cumulative,axis=0))
plt.title("cumulative sales accross all restaurant")
plt.xlabel("Years")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```
