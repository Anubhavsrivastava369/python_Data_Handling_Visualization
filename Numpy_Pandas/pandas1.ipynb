### Pandas Dataset

```python
import pandas as pd
import numpy as np


arr = np.arange(0,20).reshape(4,5)
print(arr)
```

## Create Dataset

```python
df = pd.DataFrame(data=arr,index=["Row1","Row2","Row3","Row4"],columns=["Col1","Col2","Col3","Col4","Col5",],)
print(df)
```

## Datatype

```python
data=('col1,col2,col3\n'
     'x,y,z\n'
     'a,b,c\n'
     'q,w,e\n'
     '1,2')
df=pd.read_csv(StringIO(data),dtype='object')

```

```python
df.info() 
df.isnull().sum()
```

```python
data=('col1,col2,col3\n'
     'x,y,1\n'
     'a,b,2\n'
     'q,w,3\n'
     '1,2,4')
df=pd.read_csv(StringIO(data),dtype={'col1':str,'col2':str,'col3':float})
```

```python
df.info() 
df.isnull().sum()
```
